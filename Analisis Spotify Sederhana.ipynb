```python
# Mengubah kolom 'streams' menjadi numerik, mengabaikan error
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')

# Mengambil 10 lagu dengan streams terbanyak
top_spotify_streams = df[['track_name', 'artist(s)_name', 'streams']].sort_values(by='streams', ascending=False).head(10)

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x=top_spotify_streams['streams'], y=top_spotify_streams['track_name'], palette='viridis', hue=top_spotify_streams['track_name'], dodge=False)

# Menambahkan label dan judul
plt.xlabel('Streams (in billions)')
plt.ylabel('Judul Lagu')
plt.title('10 Lagu Teratas dengan Streaming Terbanyak di Spotify')

# Menampilkan plot
plt.show()
```

```python
# Plot histogram untuk Valence
plt.figure(figsize=(10, 6))
sns.histplot(df['valence_%'], bins=20, kde=True, color='green')
plt.xlabel('Valence (%)')
plt.ylabel('Frequency')
plt.title('Distribution of Valence')
plt.show()
```

```python
# Rata-rata valence untuk setiap tahun
average_danceability_by_year = df.groupby('released_year')['valence_%'].mean()

# Plot trends valence
plt.figure(figsize=(10, 6))
sns.lineplot(x=average_danceability_by_year.index, y=average_danceability_by_year.values, color='blue')
plt.xlabel('Year')
plt.ylabel('Average Valence (%)')
plt.title('Trends in Valence Over the Years')
plt.xticks(rotation=45)
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# Plot the correlation matrix using Seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
